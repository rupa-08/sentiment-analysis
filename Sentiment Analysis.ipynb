```python
import pandas as pd

df = pd.read_csv("C:/Users/rupas/Desktop/Sentiment Analysis/sentimentdataset.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
total_null_values = df.isnull().sum()
total_null_values #there is no missing values
```

```python
#Data Cleaning

df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0']) #removing unnecessary columns
df = df.drop_duplicates()#removing duplicates rows
df
```

```python

df['Sentiment']=df['Sentiment'].str.strip()#Stripping out extra epaces
unique_sentiments = df['Sentiment'].unique()
unique_sentiments
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')#converting timestamp to consistent datetime format
```

```python
import re
#removing the unwanted characters, spaces, emojis and converting text to lower case for consistency

df['Text'] = df['Text'].apply(lambda x: re.sub(r'[^\w\s]','',str(x)).lower())
df['Text']
```

```python
#removing 
#df['Hastags'] = df['Hashtags'].apply(lambda x: x.split()if pd.notna(x) else[])
df
```

```python
#checking if Year, Month, Day and Hour datamatches with the timestamp

mismatch = df[
    (df['Year'] != df['Timestamp'].dt.year) |
    (df['Month'] != df['Timestamp'].dt.month) |
    (df['Day'] != df['Timestamp'].dt.day) |
    (df['Hour'] != df['Timestamp'].dt.hour)
]

print(mismatch)
```

```python
#sampling 10 data to see if time stamp matches with year,month,day,hour column
print(df[['Timestamp', 'Year', 'Month', 'Day', 'Hour']].sample(10))
```

```python
#Assuming timestamp column is correct and recomputing Year,month,day,hour column
df['Year']=df['Timestamp'].dt.year
df['Month']=df['Timestamp'].dt.month
df['Day']=df['Timestamp'].dt.day
df['Hour']=df['Timestamp'].dt.hour

df
```
